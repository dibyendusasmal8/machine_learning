```python
name=[]
rating=[]
n_of_review=[]
t_company=[]
hq=[]
old=[]
emplyess=[]

for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('p',class_='rating').text.strip())
    n_of_review.append(i.find('a',class_='review-count').text.strip())
    t_company.append(i.find_all('p',class_="infoEntity")[0].text.strip())
    hq.append(i.find_all('p',class_="infoEntity")[1].text.strip())
    old.append(i.find_all('p',class_="infoEntity")[2].text.strip())
    try:
        emplyess.append(i.find_all('p',class_="infoEntity")[3].text.strip())
    except:
        emplyess.append(np.nan)
    
d={'name':name,'rating':rating,'number_of_review':n_of_review,'type_of_company':t_company,'head_qater':hq,'age_of_company':old,'employ':emplyess}
df=pd.DataFrame(d)
```

```python
final=pd.DataFrame()
for j in range(0,300):
    webpage=requests.get('https://www.ambitionbox.com/list-of-companies?page={}'.format(j)).text
    soup=BeautifulSoup(webpage,'lxml')
    company=soup.find_all('div',class_='company-content-wrapper')
    name=[]
    rating=[]
    reviews=[]
    ctype=[]
    hq=[]
    how_old=[]
    no_of_employee=[]

    for i in company:

        try:
            name.append(i.find('h2').text.strip())
        except:
            name.append(np.nan)

        try:
            rating.append(i.find('p',class_='rating').text.strip())
        except:
            rating.append(np.nan)

        try:

            reviews.append(i.find('a' , class_='review-count').text.strip())
        except:
            reviews.append(np.nan)

        try:

            ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
        except:
            ctype.append(np.nan)
        try:

            hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
        except:
            hq.append(np.nan)

        try:

            how_old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
        except:
            how_old.append(np.nan)
        try:
            no_of_employee.append(i.find_all('p',class_='infoEntity')[3].text.strip())
        except:
            no_of_employee.append(np.nan)

    df=pd.DataFrame({'name':name,
        'rating':rating,
        'reviews':reviews,
        'company_type':ctype,
        'Head_Quarters':hq,
        'Company_Age':how_old,
        'No_of_Employee':no_of_employee,
        })
  
    final=final.append(df,ignore_index=True)
```
