### 1 Aam Zindegi 

```python
# adding simple imputer to fever col
si = SimpleImputer()
x_train_fever = si.fit_transform(x_train[['fever']])

# also the test data
x_test_fever = si.fit_transform(x_test[['fever']])
x_train_fever.shape
```

```python
# Ordinalencoding --> cough
ordinal_encoding = OrdinalEncoder(categories=[['Mild','Strong']])
x_train_cough = ordinal_encoding.fit_transform(x_train[['cough']])
# also the test data
x_test_cough = ordinal_encoding.fit_transform(x_test[['cough']])

x_train_cough.shape
```

```python
# OneHotEncoding -> Gender,city
ohe = OneHotEncoder(drop='first',sparse=False)
x_train_gender_city = ohe.fit_transform(x_train[['gender','city']])
# also the test data
x_test_gender_city = ohe.fit_transform(x_test[['gender','city']])
x_train_gender_city.shape
```

```python
# extract age
x_train_age = x_train.drop(columns = ['gender','fever','cough','city']).values
# also the test data
x_test_age = x_test.drop(columns = ['gender','fever','cough','city']).values
x_train_age.shape
```

```python
x_train_transfromed = np.concatenate((x_train_age,x_train_fever,x_train_gender_city,x_train_cough),axis=1)
x_test_transfromed = np.concatenate((x_test_age,x_test_fever,x_test_gender_city,x_test_cough),axis=1)
```

## Mentos zindegi 
using column transfromer methode

```python
from sklearn.compose import ColumnTransformer
```

```python
transfrmer = ColumnTransformer(transformers=[
   ('tnf1',SimpleImputer(),['fever']),
   ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
   ('tnf3',OneHotEncoder(sparse=False,drop = 'first'),['gender','city'])
],remainder='passthrough')
```

```python
ct=transfrmer.fit_transform(x_train)
```
